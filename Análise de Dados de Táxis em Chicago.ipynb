````python
# Carregar e Preparar os Dados para o Teste

from scipy import stats

# Carregando o arquivo com os dados das viagens
df_trips = pd.read_csv('/datasets/project_sql_result_07.csv')


print("--- Informações Iniciais do DataFrame de Viagens ---")
df_trips.info()

print("\n--- Amostra dos Dados ---")
print(df_trips.head())

# Verificando os valores únicos na coluna de condições climáticas
print(df_trips['weather_conditions'].unique())
````

<div class="alert alert-block alert-danger">
<b> Comentário do revisor: </b> <a class="tocSkip"></a>
<s>
    
Aqui você precisa calcular a variância para dias chuvosos e dias não chuvoses para definir o argumento `equal_var`. Você precisa usar o resultado dessa análise para definir que as variâncias são diferentes (argumento `equal_var=False`)

    
<br>
    
**Sugestão:** Você pode fazer o teste no código ou apenas imprimir as variâncias e concluir se são iguais ou diferentes. Essa etapa é importante para satisfazer a premissa do teste-t

````python
# Pré-processamento dos Dados

# Converte a coluna start_ts para o formato de data e hora
df_trips['start_ts'] = pd.to_datetime(df_trips['start_ts'])

# Extrai o dia da semana
saturday_trips = df_trips[df_trips['start_ts'].dt.day_name() == 'Saturday'].copy()

# CORREÇÃO: Usando 'Bad' para representar os dias chuvosos
bad_weather_duration = saturday_trips[saturday_trips['weather_conditions'] == 'Bad']['duration_seconds']
good_weather_duration = saturday_trips[saturday_trips['weather_conditions'] == 'Good']['duration_seconds']

# Verificação das Variância
print("Verificação de Variâncias")

var_bad_weather = bad_weather_duration.var()
var_good_weather = good_weather_duration.var()

print(f"Variância em sábados com tempo ruim: {var_bad_weather:.2f}")
print(f"Variância em sábados com tempo bom: {var_good_weather:.2f}")

# Justificativa para o parâmetro equal_var
print("\nConclusão da verificação: Como os valores das variâncias são consideravelmente diferentes,")
print("usei o argumento 'equal_var=False' no teste-t para maior precisão.\n")
````

<div class="alert alert-block alert-danger">
<b> Comentário do revisor: </b> <a class="tocSkip"></a>

<s>Veja que a amostra para `rainy` está vazia, o que não é esperado. Como sugestão, veja se o filtro está sendo feito pelo valor correto. O esperado é ter valores nas duas amostras (chuvo e não-chuvoso).</s>

</div>

````python
# Executando o Teste de Hipótese
alpha = 0.05 

results = stats.ttest_ind(
    bad_weather_duration,
    good_weather_duration,
    equal_var=False
)

print(f"P-valor (p-value): {results.pvalue}")

# Interpretando o resultado
if results.pvalue < alpha:
    print("\nHipótese nula rejeitada.")
    print("Conclusão: Existem evidências estatísticas para afirmar que a duração média das viagens muda em sábados com tempo ruim.")
else:
    print("\nNão foi possível rejeitar a hipótese nula.")
    print("Conclusão: Não há evidências estatísticas suficientes para dizer que a duração média das viagens muda em sábados com tempo ruim.")
````
